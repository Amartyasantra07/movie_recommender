```python
df['tags'] = df['genre'] + df['overview']
df
```

```python
df.head()
```

```python
new_df = df[['id','title','genre','overview', 'tags']]
```

```python
new_df
```

```python
new_df = new_df.drop(columns=['genre','overview'])
new_df
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=1000, stop_words='english')
```

```python
cv
```

```python
vec= cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Godfather']
```

```python
dist = sorted(list(enumerate(sim[2])),reverse=True,key= lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:10]:
    print(new_df.iloc[i[0]].title)
```

```python
def reccomend(movies):
    index = new_df[new_df['title']==movies].index[0]
    distance = sorted(list(enumerate(sim[index])),reverse=True, key=lambda vec:vec[1])
    for i in distance[0:10]:
        print(new_df.iloc[i[0]].title)
```

```python
reccomend("Iron Man")
```

```python
reccomend("Batman")
```

```python
reccomend("Dilwale")
```
